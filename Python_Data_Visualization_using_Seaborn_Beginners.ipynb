## Python Data Visualization using Seaborn - Beginners


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Scatter Plot

```python
sns.set(style="darkgrid")
tips=sns.load_dataset("tips")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker",style="time", data=tips)
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip",hue="size",palette="ch:r=-5,l=.75",data=tips)
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip",size="size",data=tips)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",size="size",sizes=(15,200),data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",hue="time",data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",hue="time",style="time",data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",hue="day",style="time",data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",size="size",data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill",y="tip",hue="size",size="size",data=tips)
plt.show()
```

```python
cmap=sns.cubehelix_palette(dark=.3,light=.8,as_cmap=True)
ax=sns.scatterplot(x="total_bill",y="tip",hue="size",size="size",sizes=(20,200),palette=cmap,legend="full",data=tips)
plt.show()
```

```python
cmap=sns.cubehelix_palette(dark=.3,light=.8,as_cmap=True)
ax=sns.scatterplot(x="total_bill",y="tip",hue="day",size="smoker",palette="Set2",data=tips)
plt.show()
```

```python
cmap=sns.cubehelix_palette(dark=.3,light=.8,as_cmap=True)
ax=sns.scatterplot(x="total_bill",y="tip",s=100, color=".2",marker="+",data=tips)
plt.show()
```

```python
iris=sns.load_dataset("iris")
ax=sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width,hue=iris.species,style=iris.species)
plt.show()
```

```python
index=pd.date_range("1 1 2000",periods=100,freq="m",name="date")
data=np.random.randn(100,4).cumsum(axis=0)
wide_df=pd.DataFrame(data,index,["a","b","c","d"])
ax=sns.scatterplot(data=wide_df)
```

## Lineplot

```python
df=pd.DataFrame(dict(time=np.arange(500),value=np.random.randn(500).cumsum()))
g=sns.relplot(x="time",y="value",kind="line",data=df)
plt.show()
```

```python
fmri=sns.load_dataset("fmri")
sns.relplot(x="timepoint",y="signal",kind="line",data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",ci=None,kind="line",data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",kind="line",ci="sd",data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",hue="event",kind="line",data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",hue="region",style="event",kind="line",data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",hue="region",style="event",dashes=False,markers=True,kind="line",data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",hue="event",style="event",kind="line",data=fmri)
plt.show()
```

```python
sns.relplot(x="timepoint",y="signal",hue="region",units="subject",estimator=None,kind="line",data=fmri.query("event == 'stim'"))
plt.show()
```

```python
dtos=sns.load_dataset("dots").query("align == 'dots'")
sns.relplot(x="time",y="firing_rate", hue="coherence",style="choice",kind="line",data=dtos)
plt.show()
```
